```python
from bs4 import BeautifulSoup
import os
path = "../wiki-fin/fi/articles/"

content_start = "<!-- start content -->"
content_end = "<!-- end content -->"
datapath = "./data/"
running_num = 0
for root,d_names,f_names in os.walk(path):
    if running_num > 100000:
        break
    for name in f_names:
        #running_num = running_num + 1
        try:
            full_path = root + '/' + name
            if 'jpg' not in full_path and 'JPG' not in full_path and 'png' not in full_path and 'gif' not in full_path and 'Kuva' not in full_path:
                if 'Keskustelu' not in full_path:
                    if 'Luokka' not in full_path:
                        if 'Käyttäjä' not in full_path:
                            if 'Malline' not in full_path:
                                file = open(full_path, "r")

                                html_str = file.read()

                                content_start_index = html_str.find(content_start)

                                content_end_index = html_str.find(content_end)
                                content_end_index


                                if content_start_index != -1:

                                    html_content = html_str[content_start_index + 22:content_end_index]
                                    cleantext = BeautifulSoup(html_content, "lxml").text
                                    cleantext.replace('\n', ' ')
                                    ' '.join(cleantext.split())

                                    outfile = open(datapath + str(running_num), "w")
                                    outfile.write(cleantext)
                                    outfile.close()
                                    running_num = running_num + 1

                                #else:
                                    ##print('NO CONTENT')

                                file.close()
                            #else:
                                #print('malline found')
                        #else:
                            #print('käyttäjä found')
                    #else:
                        #print('luokka found')
                #else:
                    #print('keskustelu found')
            #else:
                #print('jpg found')
        except:
            print('Error occured..')
            
```
